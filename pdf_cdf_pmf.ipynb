#  All types of distributions

## PDF (Probability Density Function):

Definition: The Probability Density Function (PDF) is used for continuous random variables. It describes the likelihood of the random variable taking on a specific value. In mathematical terms, it's the derivative of the Cumulative Distribution Function (CDF).

Python Code Example:

To illustrate the PDF of a continuous distribution, let's consider the standard normal distribution (Gaussian distribution).

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

mean = 0
std_dev = 1
x = np.linspace(-5, 5, 100)
pdf = norm.pdf(x, mean, std_dev)

plt.plot(x, pdf)
plt.title("PDF of a Standard Normal Distribution")
plt.xlabel("X")
plt.ylabel("PDF")
plt.show()
```

## CDF (Cumulative Distribution Function):

Definition: The Cumulative Distribution Function (CDF) is used for both continuous and discrete random variables. It provides the probability that the random variable is less than or equal to a specific value.

Python Code Example:

To illustrate the CDF of a continuous distribution, let's use the same standard normal distribution as in the previous example.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

mean = 0
std_dev = 1
x = np.linspace(-5, 5, 100)
cdf = norm.cdf(x, mean, std_dev)

plt.plot(x, cdf)
plt.title("CDF of a Standard Normal Distribution")
plt.xlabel("X")
plt.ylabel("CDF")
plt.show()
```

## PMF (Probability Mass Function):

Definition: The Probability Mass Function (PMF) is used for discrete random variables. It provides the probability that the random variable takes on a specific value.

Python Code Example:

To illustrate the PMF of a discrete distribution, let's consider a simple example of a fair six-sided die roll.

```python
import numpy as np
import matplotlib.pyplot as plt

outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

plt.bar(outcomes, probabilities)
plt.title("PMF of a Fair Six-sided Die")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

### 1.Normal/Gaussian Distribution:

A normal distribution, also known as a Gaussian distribution, is a probability distribution that is symmetric and bell-shaped. It is characterized by two parameters: the mean (μ) and the standard deviation (σ). 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the normal distribution
mean = 0
std_dev = 1

# Generate random samples from the normal distribution
num_samples = 1000
samples = np.random.normal(mean, std_dev, num_samples)

# Plot a histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Plot the probability density function
x = np.linspace(-5, 5, 100)
pdf = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-((x - mean)**2) / (2 * std_dev**2))
plt.plot(x, pdf, 'r--', lw=2)

plt.title("Normal Distribution")
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.show()
```

### 2. Standard normal distribution
A standard normal distribution, often denoted as Z, is a special case of the normal distribution where the mean (μ) is 0 and the standard deviation (σ) is 1. In mathematical terms, the probability density function (PDF) of the standard normal distribution is defined as:

f(z) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{z^2}{2}

Where:

f(z) is the probability density function of the standard normal distribution.

z is the standard score or Z-score, which measures the number of standard deviations a value is from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from the standard normal distribution
num_samples = 1000
samples = np.random.standard_normal(num_samples)

# Plot a histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Plot the probability density function
x = np.linspace(-3, 3, 100)
pdf = 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)
plt.plot(x, pdf, 'r--', lw=2)

plt.title("Standard Normal Distribution")
plt.xlabel("Z-score (Standard Deviations)")
plt.ylabel("Probability Density")
plt.show()
```

### 3. Log-Normal Distribution 
A log-normal distribution is a probability distribution of a random variable whose natural logarithm follows a normal distribution. In other words, if X is a log-normally distributed random variable, then 
ln ln(X) follows a normal distribution. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the log-normal distribution
mu = 0
sigma = 1

# Generate random samples from the log-normal distribution
num_samples = 1000
samples = np.random.lognormal(mean=mu, sigma=sigma, size=num_samples)

# Plot a histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Plot the probability density function
x = np.linspace(0.01, 10, 100)
pdf = 1 / (x * sigma * np.sqrt(2 * np.pi)) * np.exp(-((np.log(x) - mu)**2) / (2 * sigma**2))
plt.plot(x, pdf, 'r--', lw=2)

plt.title("Log-Normal Distribution")
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.xscale('log')  # Use a log-scale for the x-axis
plt.show()
```

### 4. Power Law Distribution
A power law distribution is a type of probability distribution where a few items have significantly higher frequencies or values compared to the majority of items. In a power law distribution, the relationship between the frequency (or value) of an item and its rank follows a power law, which is a mathematical relationship represented as 

P(X=x) is the probability of an item having a value x, and α is the exponent that determines the shape of the distribution.

In a power law distribution, smaller values have a higher probability of occurrence, and larger values have a lower probability. This distribution is characterized by a heavy tail, indicating that a few extreme values can have a significant impact on the overall distribution.

Power law distributions are common in various natural and social phenomena, including income distributions, city populations, website traffic, and word frequencies in texts.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters of the power law distribution
alpha = 2.5  # The exponent of the power law

# Generate random samples from a power law distribution
num_samples = 1000
samples = np.random.pareto(alpha, num_samples)

# Plot a histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Plot the probability density function
x = np.linspace(1, max(samples), 100)
pdf = (alpha - 1) / x**alpha
plt.plot(x, pdf, 'r--', lw=2)

plt.title("Power Law Distribution")
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.yscale('log')  # Use a log-scale for the y-axis
plt.xscale('log')  # Use a log-scale for the x-axis
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

# Define parameters of the power law distribution
alpha = 2.5  # The exponent of the power law

# Generate random samples from a power law distribution
num_samples = 1000
samples = pareto.rvs(alpha, size=num_samples)

# Plot a histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Plot the probability density function
x = np.linspace(1, max(samples), 100)
pdf = pareto.pdf(x, alpha)
plt.plot(x, pdf, 'r--', lw=2)

plt.title("Power Law Distribution")
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.yscale('log')  # Use a log-scale for the y-axis
plt.xscale('log')  # Use a log-scale for the x-axis
plt.show()
```

### 5. Bernoulli Distribution
A Bernoulli distribution is a discrete probability distribution that models the outcome of a binary event, which can have one of two possible outcomes: success (usually denoted as 1) or failure (usually denoted as 0). The distribution is characterized by a single parameter,p, which represents the probability of success.

```python
import numpy as np

# Define the probability of success (p)
p = 0.3  # Example: 30% chance of success

# Generate random samples from a Bernoulli distribution
num_samples = 10
samples = np.random.choice([0, 1], size=num_samples, p=[1 - p, p])

# Print the generated samples
print("Generated samples:", samples)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the probability of success (p)
p = 0.3  # Example: 30% chance of success

# Generate random samples from a Bernoulli distribution
num_samples = 1000
samples = np.random.choice([0, 1], size=num_samples, p=[1 - p, p])

# Create a bar plot to visualize the distribution
unique, counts = np.unique(samples, return_counts=True)
plt.bar(unique, counts, tick_label=["Failure (0)", "Success (1)"])

# Add labels and title to the plot
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution")

# Show the plot
plt.show()
```

### 6. Binomial Distribution
A binomial distribution is a discrete probability distribution that models the number of successes (usually denoted as "x") in a fixed number of independent Bernoulli trials, where each trial can have one of two possible outcomes: success or failure. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the binomial distribution
n = 10  # Number of trials
p = 0.3  # Probability of success

# Generate random samples from a binomial distribution
num_samples = 1000
samples = np.random.binomial(n, p, num_samples)

# Create a bar plot to visualize the distribution
unique, counts = np.unique(samples, return_counts=True)
plt.bar(unique, counts)

# Add labels and title to the plot
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title(f"Binomial Distribution (n={n}, p={p})")

# Show the plot
plt.show()
```

### 7. Poisson distribution
A Poisson distribution is a discrete probability distribution that describes the number of events occurring in a fixed interval of time or space when the events happen with a known constant mean rate and are independent of the time since the last event. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter of the Poisson distribution (average rate)
lambda_ = 3  # Example: Average rate of 3 events

# Generate random samples from a Poisson distribution
num_samples = 1000
samples = np.random.poisson(lambda_, num_samples)

# Create a bar plot to visualize the distribution
unique, counts = np.unique(samples, return_counts=True)
plt.bar(unique, counts)

# Add labels and title to the plot
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.title(f"Poisson Distribution (λ={lambda_})")

# Show the plot
plt.show()
```

### 8. Uniform distribution (pdf)
A uniform distribution is a continuous probability distribution that has a constant probability density, meaning that all values within a given interval are equally likely.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval [a, b]
a = 1
b = 5

# Generate random samples from a uniform distribution
num_samples = 1000
samples = np.random.uniform(a, b, num_samples)

# Create a histogram to visualize the distribution
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Add labels and title to the plot
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.title(f"Uniform Distribution [{a}, {b}]")

# Show the plot
plt.show()
```

In this example, we specify the interval [a,b] and generate random samples from a uniform distribution using np.random.uniform. We then create a histogram to visualize the distribution of the generated samples within the specified interval.

The plot demonstrates the uniform distribution property, where values are equally likely to fall within the interval 
[a,b]. You can adjust the interval and the number of samples as needed for your specific application.

### 8. Uniform Distribution(pmf)

In the case of a continuous uniform distribution, it's more appropriate to talk about the probability density function (PDF) rather than the probability mass function (PMF). The probability density function represents the likelihood of a continuous random variable taking on a specific value within a given interval. The PDF describes the relative likelihood of different values in the interval, and the total area under the PDF curve over the interval is equal to 1.

However, if you're working with a discrete uniform distribution, where there are a finite number of equally likely outcomes, you can use the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval [a, b]
a = 1
b = 5

# Generate random samples from a uniform distribution
num_samples = 1000
samples = np.random.uniform(a, b, num_samples)

# Create a histogram to visualize the PDF
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Add labels and title to the plot
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.title(f"Uniform Distribution [{a}, {b}] (PDF)")

# Show the plot
plt.show()
```

In this code, we use NumPy to generate random samples from a continuous uniform distribution within the interval [a,b]. We then create a histogram to visualize the probability density function (PDF) of the distribution, showing the relative likelihood of different values within the interval. Remember that for continuous distributions, the probability at any single point is zero, so we use density for visualization.

If you were working with a discrete uniform distribution (with a finite number of outcomes), you would use the probability mass function (PMF) to describe the probabilities associated with each outcome. However, for a continuous uniform distribution, you visualize the PDF.

### 9. Exponential Distribution
The exponential distribution is a continuous probability distribution that models the time between events in a Poisson process, where events occur at a constant average rate. It is often used to describe the time it takes for a continuous process to transition from one state to another.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the rate parameter (lambda)
lambda_ = 0.5  # Example: Average rate of 0.5 events per unit time

# Generate random samples from an exponential distribution
num_samples = 1000
samples = np.random.exponential(scale=1/lambda_, size=num_samples)

# Create a histogram to visualize the distribution
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Create the theoretical probability density function
x = np.linspace(0, max(samples), 100)
pdf = lambda_ * np.exp(-lambda_ * x)
plt.plot(x, pdf, 'r--', lw=2)

# Add labels and title to the plot
plt.xlabel("X-axis (Time between events)")
plt.ylabel("Probability Density")
plt.title(f"Exponential Distribution (λ={lambda_})")

# Show the plot
plt.show()
```

In this example, we specify the rate parameter (λ) and generate random samples from an exponential distribution using np.random.exponential. We also create a histogram to visualize the distribution of the generated samples and overlay the theoretical probability density function (PDF) on the plot.

The plot shows how the time between events is distributed, with the exponential distribution's characteristic decay in probability as the time between events increases. You can adjust the rate parameter (λ) and the number of samples as needed for your specific application.

### 10. CHI square distribution
The chi-square distribution is a continuous probability distribution that arises in various statistical tests and estimations, particularly in the context of hypothesis testing and confidence intervals. It's used to model the distribution of the sum of squared independent standard normal random variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the degrees of freedom (k)
k = 3  # Example: Degrees of freedom

# Generate random samples from a chi-square distribution
num_samples = 1000
samples = np.random.chisquare(df=k, size=num_samples)

# Create a histogram to visualize the distribution
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Create the theoretical probability density function
x = np.linspace(0, max(samples), 100)
pdf = (1/(2**(k/2) * np.math.gamma(k/2))) * x**((k/2) - 1) * np.exp(-x/2)
plt.plot(x, pdf, 'r--', lw=2)

# Add labels and title to the plot
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.title(f"Chi-Square Distribution (k={k})")

# Show the plot
plt.show()
```

### 11. F- distribution

The F-distribution, also known as the Fisher-Snedecor distribution, is a continuous probability distribution used in statistical analysis, particularly in the context of hypothesis testing and comparing the variances of two or more populations. It arises in scenarios such as analysis of variance (ANOVA) and regression analysis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the degrees of freedom parameters (d1 and d2)
d1 = 5  # Example: Degrees of freedom parameter 1
d2 = 10  # Example: Degrees of freedom parameter 2

# Generate random samples from an F-distribution
num_samples = 1000
samples = np.random.f(d1, d2, size=num_samples)

# Create a histogram to visualize the distribution
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')

# Create the theoretical probability density function
x = np.linspace(0, max(samples), 100)
pdf = (np.math.gamma((d1 + d2) / 2) / (np.math.gamma(d1 / 2) * np.math.gamma(d2 / 2))) * (d1 / d2)**(d1 / 2) * x**(d1 / 2 - 1) * (1 + (d1 * x) / d2)**(-(d1 + d2) / 2)
plt.plot(x, pdf, 'r--', lw=2)

# Add labels and title to the plot
plt.xlabel("X-axis")
plt.ylabel("Probability Density")
plt.title(f"F-Distribution (d1={d1}, d2={d2})")

# Show the plot
plt.show()
```

In this example, we specify the degrees of freedom parameters (d1,d2) and generate random samples from an F-distribution using np.random.f. We also create a histogram to visualize the distribution of the generated samples and overlay the theoretical probability density function (PDF) on the plot. The plot illustrates the characteristic shape of the F-distribution for the given degrees of freedom. You can adjust d1,d2, and the number of samples as needed for your specific analysis.
