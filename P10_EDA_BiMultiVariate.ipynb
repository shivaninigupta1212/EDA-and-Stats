```python
cross_tab_prop = pd.crosstab(index = df['Embarked'], columns = df['Sex'], normalize = "index")
cross_tab_prop
```

```python
cross_tab_prop.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 6))

plt.legend(loc="upper right", ncol=2)
plt.xlabel("Embarked")
plt.ylabel("Sex")
plt.show()
```

```python
cross_tab_prop.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 6))

plt.legend(loc="upper right", ncol=2)
plt.xlabel("Embarked")
plt.ylabel("Sex")

for n, x in enumerate([*cross_tab_prop.index.values]):
    for l in cross_tab_prop.loc[x]:

        plt.text(x=n,
                 y=l,
                 s=f'{np.round(l * 100, 1)}%',
                 color="black",
                 fontsize=12,
                 fontweight="bold")

plt.show()
```

```python
col_counts = df.groupby(['Embarked', 'Sex']).size()
col_counts = col_counts.reset_index(name = 'count')
col_counts = col_counts.pivot(index = 'Sex', columns = 'Embarked', values = 'count')

sns.heatmap(col_counts, annot = True, fmt = 'd')
```
