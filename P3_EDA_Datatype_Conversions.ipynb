```python
import pandas as pd

url = "https://raw.githubusercontent.com/bprasad26/lwd/master/data/ICICIBANK.NS.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.dtypes
```

```python
# Method 1 (using pd.to_datetime)
df['Date'] = pd.to_datetime(df['Date'])
df['Date']
```

```python
# Method 2 (using astype)
df['Date'] = df['Date'].astype('datetime64[ns]')
df['Date']
```

```python
# Method 3 (using apply function)
from datetime import datetime
df['Date'].apply(lambda val: datetime.strptime(val, "%Y-%m-%d"))
```
